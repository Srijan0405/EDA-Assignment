```python
#Task2 analysis
# üì¶ Step 1: Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("netflix_customer_churn.csv")
sns.set(style="whitegrid")
```

```python
#Plot distributions for numerical columns
numerical_cols = ['age', 'watch_hours', 'monthly_fee', 'churned']

for col in numerical_cols:
    plt.figure(figsize=(6, 4))
    sns.histplot(data=df, x=col, kde=True, bins=30, color='skyblue')
    plt.title(f'Distribution of {col.capitalize()}', fontsize=14)
    plt.xlabel(col.capitalize())
    plt.ylabel('Frequency')
    plt.tight_layout()
    plt.show()
```

```python
#Plot count plots for categorical columns
categorical_cols = [
    'subscription_type', 'gender', 'region',
    'device', 'payment_method', 'favorite_genre'
]

for col in categorical_cols:
    plt.figure(figsize=(8, 4))
    sns.countplot(data=df, x=col, order=df[col].value_counts().index, palette="Set2")
    plt.title(f'Count Plot of {col.replace("_", " ").capitalize()}', fontsize=14)
    plt.xlabel(col.replace("_", " ").capitalize())
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
#Bivarte analysis
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
categories = ['subscription_type', 'region', 'device']
for col in categories:
    plt.figure(figsize=(12, 4))
```

```python
#Average watch hours
plt.subplot(1, 2, 1)
sns.barplot(x=col, y='watch_hours', data=df, estimator='mean', palette='Blues_d')
plt.title(f'Average Watch Hours by {col.capitalize()}')
plt.xticks(rotation=45)
```

```python
#Average monthly_fee
plt.subplot(1, 2, 2)
sns.barplot(x=col, y='monthly_fee', data=df, estimator='mean', palette='Greens_d')
plt.title(f'Average Monthly Fee by {col.capitalize()}')
plt.xticks(rotation=45)
```

```python
plt.tight_layout()
plt.show()
```

```python
#Plot average avg_watch_time_per_day by favorite_genre
plt.figure(figsize=(10, 5))
sns.barplot(x='favorite_genre', y='avg_watch_time_per_day', data=df, estimator='mean', palette='pastel')
plt.title('Average Daily Watch Time by Favorite Genre')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Average churn_rate
churn_factors = ['gender', 'region', 'subscription_type', 'payment_method']

for col in churn_factors:
    plt.figure(figsize=(7, 4))
    sns.barplot(x=col, y='churned', data=df, estimator='mean', palette='Set2')
    plt.title(f'Churn Rate by {col.replace("_", " ").capitalize()}')
    plt.ylabel('Churn Rate (0 = No, 1 = Yes)')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
#Corelation analysis
plt.figure(figsize=(10, 6))
numerical_cols = df.select_dtypes(include='number')
```

```python
#Look for potential predictors of churn, e.g., watch_hours, last_login_days, monthly_fee.
corr_matrix = numerical_cols.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```

```python
Watch Hours vs. Churn
Customers with lower watch hours tend to churn more frequently.
This suggests engagement is a key predictor of churn.

Subscription Type vs. Churn
Basic users have a higher churn rate compared to Premium users.
Premium users are more likely to retain their subscription.

Region vs. Average Watch Hours
The South region shows the highest average watch hours, followed by North.
These regions may have more engaged users.

Payment Method vs. Churn
Users paying via credit card churn less than those using wallets or UPI.
Auto-renewal might be helping here.

Device Usage Patterns
Users on smart TVs and consoles tend to have higher watch hours but lower churn.
These are typically shared family accounts.

Favorite Genre vs. Engagement
Users whose favorite genre is Drama or Action tend to watch more hours daily than others.
This can guide content recommendations.

Monthly Fee vs. Churn
No strong direct correlation between monthly fee and churn was found, indicating price alone isn‚Äôt driving churn.
```
